Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = 
"timepoint" and y = "signal" for different events and regions. 

Note:  timepoint, signal, event, and region are columns in the fmri dataset.



```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

############# Main Section ############
# loading dataset using seaborn
df = sns.load_dataset('fmri')
df
```

```python
# lineplotin seaborn
sns.lineplot(x = "timepoint" , y = "signal" ,data = df)
# to show
plt.show()
```

```python
# lineplotin seaborn with hue
sns.lineplot(x = "timepoint" , y = "signal" ,data = df,hue = 'region')
# to show
plt.show()
```

```python
# lineplotin seaborn with hue and style
sns.lineplot(x = "timepoint" , y = "signal" ,data = df,hue = 'region',style = 'event')
# to show
plt.show()
```
