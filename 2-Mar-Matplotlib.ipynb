Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of 
Matplotlib.



Matplotlib is a popular data visualization library in Python. It provides a wide range of tools to create static, animated, and interactive plots in Python. Matplotlib is widely used for visualizing data in the fields of science, engineering, finance, and many others.

Some of the reasons why Matplotlib is used are:

It is a powerful and flexible library that provides a wide range of visualization options.
It is easy to learn and use.
It integrates well with other scientific computing libraries in Python such as NumPy, Pandas, and Scikit-Learn.
It has a large and active community, which means that there are many resources and examples available online.
Five plots that can be plotted using the Pyplot module of Matplotlib are:

Line Plot:
A line plot is a type of plot that displays data as a series of points connected by straight lines.

Scatter Plot:
A scatter plot is a type of plot that displays data as a collection of points.

Histogram:
A histogram is a type of plot that displays the distribution of a dataset.

Bar Plot: 
A bar plot is a type of plot that displays the frequency or distribution of categorical data.

Box Plot: 
A box plot is a type of plot that displays the distribution of a dataset based on the five-number summary (minimum, first quartile, median, third quartile, and maximum).

pyplot. subplots method provides a way to plot multiple plots on a single figure. Given the number of rows and columns , it returns a tuple ( fig , ax ), giving a single figure fig with an array of axes ax 

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
```

```python
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([ 50, 20, 40, 20, 60,70])
```

```python
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
```

```python
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig,a =  plt.subplots(2,2)
x = np.arange(1,5)
a[0][0].plot(x1,y1)
a[0][0].set_title('x1 vs y1')
a[0][1].plot(x2,y2)
a[0][1].set_title('x2 vs y2')
a[1][0].plot(x3,y3)
a[1][0].set_title('x3 vs y3')
a[1][1].plot(x4,y4)
a[1][1].set_title('x4 vs y4')
plt.show()
```

```python
# subplot
ax1 = plt.subplot(1, 1,1)
ax1.plot(x1, y1)
ax1.plot(x2, y2)
ax1.plot(x3, y3)
ax1.plot(x4, y4)
plt.show()
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000]) 
```

```python
plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company vs Profit")
```

```python
plt.barh(company,profit)
plt.ylabel("Company")
plt.xlabel("Profit")
plt.title("Company vs Profit")
```
